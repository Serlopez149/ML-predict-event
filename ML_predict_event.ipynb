Aqui seleccionamos *3000* muestras de manera aleatorias con la funcion ***sample()*** para obtener  los envios que estan en status delivered, para poder hacer una analisis estadistico mas cercano al resto de variables dentro del atributo de status, y luego lo concatenamos al dataframe que tenga el resto de variables.

```python
delivered_df = df[df['SHP_LG_SUB_STATUS'] == 'delivered'].sample(n=3000, random_state=1)
not_delivered = df[df['SHP_LG_SUB_STATUS'] != 'delivered']
df = pd.concat([delivered_df, not_delivered], axis=0)
df
```

```python
df.describe()
```

Aqui podemos observar que la variable 'delivered' es la que esta en mayor numero de sucesos, luego le sigue las devoluciones, seguido tenemos los robos, y de 4 seria las perdidas de los paquetes.

```python
plot.figure(figsize = (15,8))
ax = sns.histplot(data=df, x=df['SHP_LG_SUB_STATUS'], hue=df['SHP_LG_SUB_STATUS'])
```

En esta oportunidad visualizamos la distancia calculada antes de comenzar el trayecto y la mayoria se puede encontrar en distancias cortas.

```python
plot.figure(figsize = (15,8))
distance = df[df['SHP_LG_CALCULATED_DISTANCE'] < 1000]
sns.kdeplot(data=distance, x=distance['SHP_LG_CALCULATED_DISTANCE'])
```

al igual que los costos, tuvimos que cortar lo outliers que representaban costos muy elevados en los envios que estan representados en dolares, sin embargo la tendendicia esta mas entre los 0 USD hasta los 50 USD la representacion mas importante de los envios.

```python
plot.figure(figsize = (15,8))
df = df[df['SHP_ORDER_COST_USD'] < 200]
sns.histplot(data=df, x=df['SHP_ORDER_COST_USD'] , kde=True)
```
